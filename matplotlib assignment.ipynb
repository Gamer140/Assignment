```python
#Answer 2:
import numpy as np
import matplotlib.pyplot as plt 
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
```

```python
#Answer 3: subplot is used to draw multiple figures in single figure
import numpy as np
import matplotlib.pyplot as plt 

x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x1,y1)
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x2,y2)
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x3,y3)
plt.show()



```

```python
#Answer 5: A box plot is a special type of diagram that shows the quartiles in a box and the line extending from the lowest to the highest value.
#Box plots provide a quick visual summary of the variability of values in a dataset.
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.subplot(2,1,1)
plt.boxplot(box1)
plt.subplot(2,1,2)
plt.boxplot(box2)
```
