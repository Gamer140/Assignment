```python
#Answer 1: read_csv(): This function is used to read data from a CSV file and create a DataFrame.
#head(): This function is used to display the first n rows of a DataFrame. By default, it displays the first 5 rows.
import pandas as pd

# Reading a CSV file
data = pd.read_csv('student.csv')
print(data.head())
#info(): This function provides a summary of a DataFrame, including the data types, number of non-null values, and memory usage.
#python
data.info()
#describe(): This function computes various summary statistics of numerical columns in a DataFrame, including count, mean, standard deviation, minimum, quartiles, and maximum.
print(data.describe())
#groupby(): This function is used for grouping rows of a DataFrame based on one or more columns and performing aggregate functions on them.
df2 = pd.DataFrame({'id': [1,2,3,4,5],
                    'Age': [25, 35, 45,55,65]})
merged_data = pd.merge(data, df2,on='id')
print(merged_data)

```

```python
#Answer 2:
import pandas as pd
a = pd.DataFrame({'A':[12,34,345,4,54,56],'B':["a",'b','c','d','e','f'],'C':[0,1,0,0,1,1]})
a
def reindex_dataframe(df):
    df = df.reset_index(drop=True)
    df.index = pd.RangeIndex(start=1, stop=len(df), step=2)
    return df

new_df = reindex_dataframe(a)
new_df
```

```python
#Answer 10:
import pandas as pd
import random
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-06-26', periods=7),
    'Sales': [random.randint(1, 100) for _ in range(7)]})
data['MovingAverage'] = data['Sales'].rolling(window=7, min_periods=1).mean()
data
```
