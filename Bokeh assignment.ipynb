```python
#Answer 2: These shapes have visual information about the corresponding set of data.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x_values = [1, 2, 3, 4, 5]
y_values = [5, 8, 3, 7, 10]

# Initialize the figure
plot = figure(title="My Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis", width=400, height=400)

# Add glyphs to the plot
plot.line(x=x_values, y=y_values, line_width=2, line_color="blue", legend_label="Line")
plot.circle(x=x_values, y=y_values, size=10, fill_color="red", line_color="black", legend_label="Circle")

# Customize the plot
plot.legend.location = "top_left"
plot.legend.title = "Glyphs"

# Show the plot in a Jupyter notebook
output_notebook()
show(plot)
```
